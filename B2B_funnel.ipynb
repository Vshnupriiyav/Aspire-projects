```python
#import liabraries
import pandas as pd
from datetime import datetime
```

```python

def get_user_journey_metrics(conn):
    # Load base tables
    users_df = pd.read_sql("SELECT * FROM user", conn)
    
    # Get latest bureau consent
    bureau_consent_df = pd.read_sql("""
        SELECT user_id, consent_type, fl_consent 
        FROM user_app_consent_master 
        WHERE id IN (
            SELECT MAX(id) 
            FROM user_app_consent_master 
            WHERE consent_type = 'bureau' 
            GROUP BY user_id
        )
    """, conn)
    
    # Get latest OTP verification
    otp_df = pd.read_sql("""
        SELECT user_id 
        FROM otp_api_partner_config 
        WHERE id IN (
            SELECT MAX(id) 
            FROM otp_api_partner_config 
            WHERE event = 'signup' 
            GROUP BY user_id
        )
    """, conn)
    
    # Get latest questionnaire responses
    questionnaire_df = pd.read_sql("""
        SELECT 
            user_id,
            CASE 
                WHEN json_extract(questions_response, '$[0].question_response[0].options[0].field_text') IS NOT NULL 
                THEN json_unquote(json_extract(questions_response, '$[0].question_response[0].options[0].field_text'))
                ELSE json_unquote(json_extract(questions_response, '$[0].options[0].option_text'))
            END as language
        FROM questionnaire_response 
        WHERE id IN (
            SELECT MAX(id) 
            FROM questionnaire_response 
            GROUP BY user_id
        )
    """, conn)
    
    # Load other required tables with latest records
    address_df = pd.read_sql("SELECT DISTINCT user_id FROM user_address", conn)
    crif_df = pd.read_sql("""
        SELECT user_id, status 
        FROM crif_stage2 
        WHERE id IN (
            SELECT MAX(id) 
            FROM crif_stage2 
            GROUP BY user_id
        )
    """, conn)
    
    bureau_check_df = pd.read_sql("""
        SELECT user_id, approved 
        FROM linedetails 
        WHERE id IN (
            SELECT MAX(id) 
            FROM linedetails 
            WHERE purpose = 'bureaucheck' 
            GROUP BY user_id
        )
    """, conn)
    
    postkyc_df = pd.read_sql("""
        SELECT user_id, approved 
        FROM linedetails 
        WHERE id IN (
            SELECT MAX(id) 
            FROM linedetails 
            WHERE purpose = 'postkyc' 
            GROUP BY user_id
        )
    """, conn)
    
    sanction_df = pd.read_sql("SELECT * FROM sanction", conn)
    
    merchant_payments_df = pd.read_sql("""
        SELECT user_id, loan_start_date, utr 
        FROM merchant_payments 
        WHERE id IN (
            SELECT MIN(id) 
            FROM merchant_payments 
            WHERE status = 'success' 
            GROUP BY user_id
        )
    """, conn)
    
    coupon_df = pd.read_sql("SELECT user_id, coupon_code FROM user_registeration_coupon_code_details", conn)
    wrongly_b2b_df = pd.read_sql("SELECT pan FROM asp_playground.wrongly_b2b_preapproved", conn)
    
    # Merge all dataframes
    result_df = users_df.merge(coupon_df, left_on='id', right_on='user_id', how='left')\
        .merge(bureau_consent_df, left_on='id', right_on='user_id', how='left')\
        .merge(otp_df, left_on='id', right_on='user_id', how='left')\
        .merge(questionnaire_df, left_on='id', right_on='user_id', how='left')\
        .merge(address_df, left_on='id', right_on='user_id', how='left')\
        .merge(crif_df, left_on='id', right_on='user_id', how='left')\
        .merge(bureau_check_df, left_on='id', right_on='user_id', how='left')\
        .merge(postkyc_df, left_on='id', right_on='user_id', how='left')\
        .merge(sanction_df, left_on='id', right_on='user_id', how='left')\
        .merge(merchant_payments_df, left_on='id', right_on='user_id', how='left')
    
    # Filter out wrongly pre-approved B2B users and specific coupon codes
    result_df = result_df[
        (~result_df['pannumber'].isin(wrongly_b2b_df['pan'])) &
        (result_df['coupon_code'].isin(['ESSMART', 'roinet', 'gramophone', 'yodacart', 'buyume']))
    ]
    
    # Convert created_at to date
    result_df['created_at'] = pd.to_datetime(result_df['created_at']).dt.strftime('%Y-%m-%d')
    
    # Group by date and coupon code with aggregations
    metrics_df = result_df.groupby(['created_at', 'coupon_code']).agg({
        'phone_number': 'count',
        'fl_consent': lambda x: (x == 1).sum(),
        'user_id_x': lambda x: x.notna().sum(),  # OTP triggered
        'is_active': lambda x: (x == 1).sum(),
        'pannumber': lambda x: ((result_df['is_active'] == 1) & result_df['pannumber'].notna()).sum(),
        'dob': lambda x: ((result_df['is_active'] == 1) & result_df['dob'].notna()).sum(),
        'gender': lambda x: ((result_df['is_active'] == 1) & result_df['gender'].notna()).sum(),
        'language': 'count',
        'email': lambda x: ((result_df['is_active'] == 1) & result_df['email'].notna()).sum(),
        'user_id_y': 'count',  # completed address
        'status': lambda x: (x == 'S11').sum(),  # Need alternate number
        'status': lambda x: (x == 'S10').sum(),  # CRIF S10 users
        'approved_x': lambda x: (x == 1).sum(),  # Bureau approved
        'approved_x': lambda x: (x == 0).sum(),  # Bureau rejected
        'approved_x': lambda x: (x == -1).sum(),  # Bureau manual review
        'approved_y': lambda x: (x == 1).sum(),  # KYC done
        'user_id': 'count',  # sanction created
        'is_enach_setup_required': lambda x: (x == 1).sum(),
        'is_enach_setup_done': lambda x: (x == 1).sum(),
        'fl_active': lambda x: (x == 1).sum(),
        'loan_start_date': lambda x: x.notna().sum(),
        'utr': lambda x: x.notna().sum()
    }).reset_index()
    
    # Rename columns to match SQL query
    metrics_df.columns = [
        'created_at', 'coupon_code', 'phone_number_entered', 'bureau_consent_given',
        'OTP_Triggered', 'otp_triggered_and_verifed', 'PAN_entered', 'DOB_entered',
        'gender_entered', 'q_language_entered', 'email_entered', 'completed_address',
        'Need_alternate_number_for_bureau', 'crif_S10_users', 'user_approved_in_bureau',
        'user_rejected_in_bureau', 'user_in_bureau_manual_review', 'kyc_done',
        'sanction_created', 'nach_required', 'nach_done', 'line_activated',
        'txn_requested', 'txn_completed'
    ]
    
    return metrics_df.sort_values(['created_at', 'coupon_code'])

# Example usage
if __name__ == "__main__":
    # Replace with your database connection
    
    from sqlalchemy import create_engine
    engine = create_engine('mysql://"vishnupriya":password@host/database')
    
    dataframe = get_user_journey_metrics(engine)
    print(dataframe.head())
```
